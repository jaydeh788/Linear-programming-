```python
import pulp
```

```python
Lp_prob = pulp.LpProblem("Q2d", pulp.LpMaximize)
```

```python
# Define decision variables 
x1 = pulp.LpVariable("x1",lowBound=0)  
x2 = pulp.LpVariable("x2",lowBound=0)
x3 = pulp.LpVariable("x3",lowBound=0)
```

```python
Lp_prob += 300*x1 + 400*x2 + 1000*x3 
```

```python
Lp_prob += x1 + x2 <= 40
Lp_prob += x2 + 2*x3 <= 30
Lp_prob += 2*x1 <= 20
Lp_prob += x1 + 2*x3 <= 10
Lp_prob += x3 + 2*x2 <= 10

Lp_prob += x1 >= 0
Lp_prob += x2 >= 0
Lp_prob += x3 >= 0
```

```python
print (Lp_prob)
```

```python
Lp_prob.solve()
pulp.LpStatus[Lp_prob.status]
```

```python
print("x1=", pulp.value(x1),
      "x2=", pulp.value(x2),
      "x3=", pulp.value(x3))
print("optimal value is z =", pulp.value(Lp_prob.objective))
```
