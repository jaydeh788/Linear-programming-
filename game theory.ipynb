```python
import pulp
```

```python
from pulp import *
```

```python
M=100
N=100
```

```python
column=[j+1 for j in range(N)]
```

```python
row=[i+1 for  i in range(M)]
```

```python
def a(i,j):
    if i ==j: return 0
    if i==j-1:return -1
    if i<j-1: return 1
    if i==j+1: return 1
    if i>j+1: return -1
```

```python
a={i:{j:a(i,j)for j in column}for i in row}
```

```python
x=["x_"+str(i) for i in column]
v= ["v"]
```

```python
prob = LpProblem("HW10", LpMaximize)
prob_vars = LpVariable.dicts("var",x+v) 
```

```python
#OBJECTIVE FUNCTION
prob += prob_vars['v']
```

```python
#CONSTRAINTS
for j in column:
        prob+= prob_vars['v']-lpSum(a[i][j]*prob_vars["x_"+str(i)] for i in row)<=0

#probability adds up to 1
prob+=lpSum(prob_vars["x_"+str(i)] for i in row)==1

#probability must be non-negative
for i in row:
    prob+=prob_vars["x_"+str(i)]>=0
```

```python
prob.solve()
print("Status:", LpStatus[prob.status])
```

```python
print("The optimal value = ", value(prob.objective))
```

```python

for i in prob.variables():
    print(i.name, "=", i.varValue)
```
